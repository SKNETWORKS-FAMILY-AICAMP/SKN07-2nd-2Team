```python
# 민혁

# 이상치 처리

# HandsetPrice: 휴대폰 가격대
# Unknown -> 소득수준 컬럼 IncomeGroup 로 그룹화해서 평균값 채워넣기

df_encoding = df_drop.copy()

price_data = df_encoding[df_encoding.HandsetPrice != 'Unknown']
price_data.loc[:,'HandsetPrice'] = price_data['HandsetPrice'].astype('int64')
handset_mean_price = price_data.groupby(['IncomeGroup'])[['HandsetPrice']].mean()

for i,y in handset_mean_price.iterrows():
    df_encoding.loc[(df_encoding.HandsetPrice == 'Unknown') & (df_encoding.IncomeGroup == i), 'HandsetPrice'] = round(y.values[0])

df_encoding['HandsetPrice'] = df_encoding['HandsetPrice'].astype('int64')
```

```python
# 재철

# 범주형 컬럼 라벨링

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

# CreditRating: 소득수준 그룹 (순서 변환)
df_encoding['IncomeGroup'] = df_encoding['IncomeGroup'].apply(lambda x: 10-x)

# CreditRating: 신용등급 (순서 변환)
credit_rating_mapping = { '1-Highest': 7, '2-High': 6, '3-Good': 5, '4-Medium': 4, '5-Low': 3, '6-VeryLow': 2, '7-Lowest': 1 }
df_encoding['CreditRating'] = df_encoding['CreditRating'].map(credit_rating_mapping)

# PrizmCode: 라이프스타일 구분 (라벨인코딩)
df_encoding['PrizmCode'] = le.fit_transform(df_encoding['PrizmCode'])  
# Occupation: 직업 (라벨인코딩)
df_encoding['Occupation'] = le.fit_transform(df_encoding['Occupation'])
# ServiceArea: 서비스 지역 (라벨인코딩)
df_encoding['ServiceArea'] = le.fit_transform(df_encoding['ServiceArea'])
```

```python
# 수민

# BIT 여부 컬럼
# No/Yes = 0/1 라벨링

col_object = df_encoding.columns[df_encoding.dtypes=='object']
col_bit = [col for col in col_object if 'Yes' in list(df_encoding[col].unique())]

# MaritalStatus: 혼인 여부 (Yes/No/Unknown)
col_bit.remove('MaritalStatus')
df_encoding['MaritalStatus'] = df_encoding['MaritalStatus'].apply(lambda x: -1 if x == 'Unknown' else x)
df_encoding['MaritalStatus'] = df_encoding['MaritalStatus'].apply(lambda x: 1 if x == 'Yes' else 0)

# Homeownership: 주택 소유 여부 (Known/Unknown)
df_encoding['Homeownership'] = df_encoding['Homeownership'].apply(lambda x: 1 if x == 'Known' else 0)

# 기타 여부 컬럼들 (Yes/No)
for col in col_bit:
    df_encoding[col] = df_encoding[col].apply(lambda x: 1 if x == 'Yes' else 0)
```

### 모델 검증 결과 함수화

```python
# 분류 모델 혼합행렬 결과 print 함수
def matrix_print(model_nm, model, y_preds, y_val):
    from sklearn.metrics import confusion_matrix, accuracy_score, recall_score, precision_score, f1_score, roc_auc_score
    print('-'*50)
    print(model_nm, '\n')
    
    print('  Confusion Matrix')
    print('  ', confusion_matrix(y_preds, y_val), '\n')

    print('  정확도: ', round(accuracy_score(y_preds, y_val), 2))
    print('  recall: ', round(recall_score(y_preds, y_val), 2))
    print('  정밀도: ', round(precision_score(y_preds, y_val), 2))
    print('  F1: ', round(f1_score(y_preds, y_val), 2), '\n')

    print('  ROC AUC: ', round(roc_auc_score(y_test, model.predict_proba(X_test)[:, 1]), 2), '\n')
```

### ML 모델 학습

```python
def ml_model_train(X_train, y_train, X_val, y_val):
    # 로지스틱 회귀
    from sklearn.linear_model import LogisticRegression
    model_lr = LogisticRegression(max_iter=1000, random_state=42)
    model_lr.fit(X_train, y_train)
    y_preds_lr = model_lr.predict(X_val)
    matrix_print('로지스틱회귀', model_lr, y_preds_lr, y_val)

    # Decision Tree
    from sklearn.tree import DecisionTreeClassifier
    model_dt = DecisionTreeClassifier(max_depth=3, random_state=42)
    model_dt.fit(X_train, y_train)
    y_preds_dt = model_dt.predict(X_val)
    matrix_print('Decision Tree', model_dt, y_preds_dt, y_val)

    # RandomForest
    from sklearn.ensemble import RandomForestClassifier
    model_rf = RandomForestClassifier(n_jobs=5, random_state=42, max_depth=20)
    model_rf.fit(X_train, y_train)
    y_preds_rf = model_rf.predict(X_val)
    matrix_print('RandomForest', model_rf, y_preds_rf, y_val)

    # LightGBM
    from lightgbm import LGBMClassifier
    model_lgb = LGBMClassifier(n_estimators=100, learning_rate=0.1, random_state=42)
    model_lgb.fit(X_train, y_train)
    y_preds_lgb = model_lgb.predict(X_val)
    matrix_print('LightGBM', model_lgb, y_preds_lgb, y_val)

    # XGBoost
    from xgboost import XGBClassifier
    model_xgb = XGBClassifier(n_estimators=100, learning_rate=0.1, use_label_encoder=False, eval_metric='logloss', random_state=42)
    model_xgb.fit(X_train, y_train)
    y_preds_xgb = model_xgb.predict(X_val)
    matrix_print('XGBoost', model_xgb, y_preds_xgb, y_val)

    return model_lr, model_dt, model_rf, model_lgb, model_xgb
```

```python
model_lr, model_dt, model_rf, model_lgb, model_xgb = ml_model_train(X_train, y_train, X_val, y_val)
```

```python
import pandas as pd
import torch
import torch.nn as nn
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from torch.utils.data import DataLoader, TensorDataset

# 데이터 정규화 (특징들을 스케일링)
ss = StandardScaler()   # 데이터 정규화를 위한 StandardScaler 사용
ss.fit(X_train)
X_train_ss = ss.transform(X_train) # 데이터를 평균 0, 표준편차 1로 변환
X_val_ss = ss.transform(X_val) # 데이터를 평균 0, 표준편차 1로 변환

# 데이터를 PyTorch 텐서 형식으로 변환
X_train_tensor = torch.tensor(X_train_ss, dtype=torch.float32) # 훈련 데이터
y_train_tensor = torch.tensor(np.array(y_train), dtype=torch.float32).view(-1, 1) # 타겟 데이터 (이진 분류이므로 2D 텐서로 변환)
X_val_tensor = torch.tensor(X_val_ss, dtype=torch.float32) # 검증 데이터
y_val_tensor = torch.tensor(np.array(y_val), dtype=torch.float32).view(-1, 1)   # 타겟 데이터(이진 분류이므로 2D 텐서로 변환)

# DataLoader로 배치 처리 (훈련 데이터, 테스트 데이터)
train_data = TensorDataset(X_train_tensor, y_train_tensor)  # 훈련 데이터셋 생성
val_data = TensorDataset(X_val_tensor, y_val_tensor) # 테스트 데이터셋 생성
train_dl = DataLoader(train_data, batch_size=64, shuffle=True)  # 훈련 데이터셋에 배치 크기 설정, 데이터를 섞음
val_dl = DataLoader(val_data, batch_size=64, shuffle=False)   # 테스트 데이터셋에 배치 크기 설정
```

```python
# 신경망 모델 정의
class tele_churn_predict(nn.Module):
    def __init__(self):
        super(tele_churn_predict, self).__init__()
        # 개의 입력 특징을 받아들이는 첫 번째 Fully Connected Layer
        self.fc1 = nn.Linear(56, 32)  # 개의 입력 특징
        self.fc2 = nn.Linear(32, 16)  # 32개의 출력에서 16개로 변환
        self.fc3 = nn.Linear(16, 1)  # 16개의 출력에서 1개의 생존 확률 값으로 변환
        self.sigmoid = nn.Sigmoid()  # 생존 확률을 0과 1 사이로 변환하기 위한 Sigmoid 함수

    def forward(self, x):
        x = nn.ReLU()(self.fc1(x))  # 첫 번째 레이어 후 ReLU 활성화 함수
        x = nn.ReLU()(self.fc2(x))  # 두 번째 레이어 후 ReLU 활성화 함수
        x = self.fc3(x)  # 세 번째 레이어
        x = self.sigmoid(x)  # 출력값을 확률로 변환
        return x

# 모델 인스턴스화
model = tele_churn_predict()
# 손실 함수와 옵티마이저 설정
loss_fn = nn.BCELoss()  # 이진 분류 문제이므로 Binary Cross Entropy 손실 함수 사용
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)  # Adam 옵티마이저 사용

# 훈련 루프
num_epochs = 20  # 훈련할 에포크 수 설정
for epoch in range(num_epochs):
    model.train()  # 모델을 훈련 모드로 설정
    total_loss = 0  # 손실을 누적할 변수
    correct = 0  # 정확도를 계산할 변수
    for X_batch, y_batch in train_dl:
        optimizer.zero_grad()  # 기울기 초기화
        output = model(X_batch)  # 순전파: 배치 데이터를 모델에 통과시켜 예측값 생성
        loss = loss_fn(output, y_batch)  # 손실 계산
        loss.backward()  # 역전파
        optimizer.step()  # 옵티마이저로 파라미터 업데이트
        total_loss += loss.item()  # 손실값 누적
        predicted = (output > 0.5).float()  # 예측값이 0.5보다 크면 생존, 작으면 사망
        correct += (predicted == y_batch).sum().item()  # 정확도 계산

    # 에포크마다 손실과 정확도 출력
    print(f"Epoch {epoch+1}/{num_epochs}, Loss: {total_loss/len(train_dl):.4f}, Accuracy: {correct/len(train_dl.dataset):.4f}")
```
